```python
import pandas as pd
import os
import matplotlib.pyplot as plt
```

Task #1:Merging 12 months of data into a single csv file

```python
df = pd.read_csv("./Sales_Data/Sales_April_2019.csv")
sales_files=[file for file in os.listdir('.\Sales_Data')]
sales_data = pd.DataFrame()
for file in sales_files:
    df=pd.read_csv('./Sales_Data/'+file)
    sales_data = pd.concat([sales_data,df])

sales_data.to_csv('all_sales_data.csv',index=False)
```

#### Clean up the data

```python
nan_df = sales_data[sales_data.isna().any(axis=1)]
nan_df.head()

sales_data = sales_data.dropna(how='all')
```

#### Find OR and delete it

```python
sales_data=sales_data[sales_data['Order Date'].str[0:2]!='Or']
```

#### Augment data with additional colums

```python
#lets use .apply()
def get_city(address):
    return address.split(',')[1]

def get_state(address):
    return address.split(',')[2].split(' ')[1]

sales_data['city']=sales_data['Purchase Address'].apply(lambda x:f"{get_city(x)} ({get_state(x)})")

sales_data['Month']=sales_data['Order Date'].str[0:2]
sales_data['Month']=sales_data['Month'].astype('int32')
sales_data['Quantity Ordered']=pd.to_numeric(sales_data['Quantity Ordered'])
sales_data['Price Each']=pd.to_numeric(sales_data['Price Each'])
sales_data['sales']=sales_data['Quantity Ordered']*sales_data['Price Each']
sales_data.head()
```

#### What was the best month for sales?How much was earned that month

```python
results = sales_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)
plt.bar (months,results['sales'])
plt.xticks(months)
plt.ylabel('Sales in USD ($)')
plt.xlabel('Month Number')
plt.show()
```

#### What city has the highest number of sales

```python
results_city = sales_data.groupby('city').sum()
results_city
```

```python


cities = [city for city, df in sales_data.groupby('city')]
plt.bar (cities,results_city['sales'])
plt.xticks(cities,rotation='vertical')
plt.ylabel('Sales in USD ($)')
plt.xlabel('Cities')
plt.show()
```

#### What time should we display advertisements to maximise sales?

```python
sales_data['Order Date']=pd.to_datetime(sales_data['Order Date'])
sales_data.head()
```

```python
sales_data['Hour']=sales_data['Order Date'].dt.hour
sales_data['Minute']=sales_data['Order Date'].dt.minute
sales_data.head()
```

```python
hours = [hour for hour, df in sales_data.groupby('Hour')]
#sales_data.groupby(['Hour']).count()
plt.figure(figsize=(5,3),dpi=150)
plt.plot(hours,sales_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel('Hour')
plt.ylabel('Number of Orders')
plt.grid()
plt.show()
```

#### What products are most often sold together?

```python
df = sales_data[sales_data['Order ID'].duplicated(keep=False)]
#df.head(20)
df['grouped']=df.groupby('Order ID')['Product'].transform(lambda x:','.join(x))

df = df[['Order ID','grouped']].drop_duplicates()
df.head()
```

```python
from itertools import combinations
from collections import Counter

#list3 = [[2, 3], [2, 3, 4]]

count = Counter()

for row in df['grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))

for key,value in count.most_common(10):
    print(key,value)
```

#### What product sold the most? Why do you think it sold the most?

```python
product_group = sales_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
products = [product for product,df in product_group]

plt.figure(figsize=(5,3),dpi=150)
plt.bar(products,quantity_ordered)
plt.ylabel('Quantity Ordered')
plt.xlabel('Product')
plt.xticks(products,rotation='vertical',size=8)
plt.show()
```

```python
prices = sales_data.groupby('Product')
prices_mean = prices.mean()['Price Each']

fig, ax = plt.subplots(figsize = (10, 5),dpi=300)
plt.title('Product and Reason')
  
# using the twinx() for creating another
# axes object for secondry y-Axis
ax2 = ax.twinx()
ax.bar(products,quantity_ordered, color = 'g')
ax2.plot(products, prices_mean, color = 'b')


# giving labels to the axises
ax.set_xlabel('Product Names', color = 'r')
ax.set_ylabel('Quantity Ordered', color = 'g')

ax.set_xticklabels(products,rotation='vertical',size=8)
# secondary y-axis label
ax2.set_ylabel('Prices in USD ($)', color = 'b')
  
# defining display layout 
plt.tight_layout()

# Save the graph
plt.savefig('Product_Quantity_Price.png',dpi=300)
# show plot
plt.show()
```
